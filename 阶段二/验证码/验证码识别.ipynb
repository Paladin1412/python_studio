```python
from tencentcloud.common import credential
from tencentcloud.common.profile.client_profile import ClientProfile
from tencentcloud.common.profile.http_profile import HttpProfile
from tencentcloud.common.exception.tencent_cloud_sdk_exception import TencentCloudSDKException 
from tencentcloud.ocr.v20181119 import ocr_client, models 
import json
import time
def picture_api(url):
    try: 
        cred = credential.Credential("AKID9vSAbneNmh0woQovnnXzEDwF0FK8lgOe", "EBgpW6FrAIhAvQjmY4dx0smBw3u5dVkj") 
        httpProfile = HttpProfile()
        httpProfile.endpoint = "ocr.tencentcloudapi.com"

        clientProfile = ClientProfile()
        clientProfile.httpProfile = httpProfile
        client = ocr_client.OcrClient(cred, "ap-guangzhou", clientProfile) 

        req = models.GeneralHandwritingOCRRequest()
        params = '{\"ImageUrl\":\"%s\"}'%(url)
        req.from_json_string(params)

        resp = client.GeneralHandwritingOCR(req) 
        result = resp.to_json_string()
        result = json.loads(result)
        result = result['TextDetections'][0]['DetectedText']
        return result

    except TencentCloudSDKException as err: 
        return ''
```
