```python
#手动登录
def login():
    options = webdriver.ChromeOptions()
    '''设置浏览器为手机显示'''
    mobile_emulation = {"deviceName":"Galaxy S5"}
    capabilities = DesiredCapabilities.CHROME
    capabilities['loggingPrefs'] = { 'browser':'ALL' }
    options = webdriver.ChromeOptions()
    options.add_experimental_option("mobileEmulation", mobile_emulation)
    options.add_experimental_option("excludeSwitches", ['enable-automation'])
    #修改windows.navigator.webdriver，防机器人识别机制，selenium自动登陆判别机制
    options.add_experimental_option('excludeSwitches', ['enable-automation'])
    drive = webdriver.Chrome(executable_path='E:\Google\Chrome\Application\chromedriver.exe',desired_capabilities=capabilities,options=options)
 	#CDP执行JavaScript 代码  重定义windows.navigator.webdriver的值
    drive.execute_cdp_cmd("Page.addScriptToEvaluateOnNewDocument", {
      "source": """
        Object.defineProperty(navigator, 'webdriver', {
          get: () => undefined
        })
      """
    })
    url = 'https://h5.ele.me/'
    drive.implicitly_wait(10)
    drive.get(url)
    my = drive.find_element_by_xpath('//*[@ubt-click="105141"]')
    my.click()
    time.sleep(1)
    signup_login = drive.find_element_by_xpath('//*[@class="profile-1_mtk"]')
    signup_login.click()
    input('请手动登陆(登陆完成后按任意键继续)...')
    drive.maximize_window() #窗口最大化
    time.sleep(2)
    return drive
web = login()
```
